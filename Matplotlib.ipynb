## labels

```python
plt.plot([1,2,3] , [2,4,6])
plt.title("this is ritesh data")
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.show()
```

## Marker

```python
# it is representation and point in graph
```

```python
# syntax : plot(x , y , marker = "" , for color (mec = "" (edge color) , mfc  "" (body color)) , for size (ms = "") )
```

```python
import numpy as np
x = np.arange(0 , 10 , 2)
y = x ** 2
plt.plot(x, y , marker = "o" , ms = '10' , mec = "r" , mfc = "g")
plt.title("this is ritesh data" )
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.show()
```

## Line

```python
# syntax : plot(x , y , ls = "" (linestyle means hows line ex. solid , dotted , dashed , dashdot ,none) , lw=""(line width) 
#  , c = "" (linecolor ex. r , g , b etc)) 
```

```python
x = np.arange(0 , 5)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
plt.plot(x, y1 ,ls = "dashdot" , lw = "5" , c = "r")
plt.plot(x, y2,ls = "solid" , lw = "5" , c = "g")
plt.plot(x, y3 ,ls = "dashed" , lw = "5" , c = "b")
plt.title("this is ritesh data" )
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.show()
```

## Grid

```python
x = np.arange(0 , 5)
y = x ** 2
plt.plot(x, y ,ls = "dashdot" , lw = "5" , c = "r")
plt.grid(axis = "both" , ls = "dotted")
plt.title("this is ritesh data" )
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.show()
```
